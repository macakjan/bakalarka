```python
import minkowski_from_gram_6D_data

# G must be sorted (diagonal elements not decreasing)
# returned [] means 'is reduced'
def minkowski_check_6D(G):
    changes = minkowski_from_gram_6D_data.changes_6D
    combs = minkowski_from_gram_6D_data.combs_6D
    diag = [G[i,i] for i in range(6)]
    cur_sum = G[0,0] + 2*G[1,0] + G[1,1]
    coprime_index = 1
    for i in range(2469):  # 2469 == len(combs)-1
        if cur_sum < diag[coprime_index]:
            return [coprime_index, i]
        j, d, coprime_index = changes[i]
        for k in range(6):
            cur_sum += 2*d*combs[i][k]*G[j, k]
        cur_sum += d*d*G[j,j]
    if cur_sum < diag[coprime_index]:
        return [coprime_index, 2469]
    return []

def minkowski_check_5D(G):
    changes = minkowski_from_gram_6D_data.changes_5D
    combs = minkowski_from_gram_6D_data.combs_5D
    diag = [G[i,i] for i in range(5)]
    cur_sum = G[0,0] + 2*G[1,0] + G[1,1]
    coprime_index = 1
    for i in range(195):  # 195 == len(combs)-1
        if cur_sum < diag[coprime_index]:
            return [coprime_index, i]
        j, d, coprime_index = changes[i]
        for k in range(5):
            cur_sum += 2*d*combs[i][k]*G[j, k]
        cur_sum += d*d*G[j,j]
    if cur_sum < diag[coprime_index]:
        return [coprime_index, 195]
    return []

def minkowski_check_4D(G):
    changes = minkowski_from_gram_6D_data.changes_4D
    combs = minkowski_from_gram_6D_data.combs_4D
    diag = [G[i,i] for i in range(4)]
    cur_sum = G[0,0] + 2*G[1,0] + G[1,1]
    coprime_index = 1
    for i in range(35):  # 35 == len(combs)-1
        if cur_sum < diag[coprime_index]:
            return [coprime_index, i]
        j, d, coprime_index = changes[i]
        for k in range(4):
            cur_sum += 2*d*combs[i][k]*G[j, k]
        cur_sum += d*d*G[j,j]
    if cur_sum < diag[coprime_index]:
        return [coprime_index, 35]
    return []

def minkowski_check_3D(G):
    changes = minkowski_from_gram_6D_data.changes_3D
    combs = minkowski_from_gram_6D_data.combs_3D
    diag = [G[i,i] for i in range(3)]
    cur_sum = G[0,0] + 2*G[1,0] + G[1,1]
    coprime_index = 1
    for i in range(9):  # 9 == len(combs)-1
        if cur_sum < diag[coprime_index]:
            return [coprime_index, i]
        j, d, coprime_index = changes[i]
        for k in range(3):
            cur_sum += 2*d*combs[i][k]*G[j, k]
        cur_sum += d*d*G[j,j]
    if cur_sum < diag[coprime_index]:
        return [coprime_index, 9]
    return []

# returned 1 means the sum is smaller, -1 the difference is smaller, 0 means reduced
def minkowski_check_2D(G):
    cur_sum = G[0,0] + 2*G[1,0] + G[1,1]
    if cur_sum < G[1,1]:
        return 1
    if cur_sum - 4*G[1,0] < G[1,1]:
        return -1
    return 0
```
